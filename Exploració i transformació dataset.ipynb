```python

# Definir funció de macro-gèneres (10 grups)
def assign_macro_genre(genre):
    g = genre.lower()

    # 1. Pop
    if 'pop' in g:
        return 'Pop'

    # 2. Rock
    elif g in [
        'rock', 'alt-rock', 'hard-rock', 'punk-rock', 'rock-n-roll',
        'psych-rock', 'grunge', 'indie', 'indie-pop', 'power-pop'
    ]:
        return 'Rock'

    # 3. Metal
    elif g in [
        'metal', 'black-metal', 'death-metal',
        'heavy-metal', 'metalcore', 'grindcore'
    ]:
        return 'Metal'

    # 4. Electrónica
    elif g in [
        'edm', 'electro', 'electronic', 'house', 'techno',
        'minimal-techno', 'progressive-house', 'dubstep',
        'trance', 'breakbeat', 'chicago-house', 'detroit-techno'
    ]:
        return 'Electrónica'

    # 5. Hip-hop / R&B
    elif g in ['hip-hop', 'r-n-b', 'funk', 'garage']:
        return 'Hip-hop/R&B'

    # 6. Llatí
    elif g in [
        'latin', 'latino', 'salsa', 'pagode', 'brazil',
        'forro', 'sertanejo', 'samba', 'mpb'
    ]:
        return 'Llatí'

    # 7. Jazz / Clàssica
    elif g in ['jazz', 'classical', 'opera']:
        return 'Jazz/Clàssica'

    # 8. Ambient / Chill
    elif g in ['ambient', 'chill', 'sleep', 'study', 'new-age', 'piano']:
        return 'Ambient/Chill'

    # 9. Infantil
    elif g in ['children', 'kids', 'disney']:
        return 'Infantil'

    # 10. World / Regional
    elif g in [
        'world-music', 'indian', 'iranian', 'turkish',
        'swedish', 'german', 'french', 'spanish', 'tango'
    ]:
        return 'World/Regional'

    # Altres (residual)
    else:
        return 'Altres'

# Crear columna de macro-gènere
df = df.copy()
df['macro_genre'] = df['track_genre'].apply(assign_macro_genre)

# Comprovacions
# Quants macro-gèneres tenim
print(df['macro_genre'].nunique())

# Repartiment de cançons per macro-gènere
print(df['macro_genre'].value_counts())

# Eliminar "altres"
df_10_genres = df[df['macro_genre'] != 'Altres']

# Guardar el nou dataset
df_10_genres.to_csv("dataset_10_generes.csv", index=False)

print("Dataset guardat com: dataset_10_generes.csv")
```
